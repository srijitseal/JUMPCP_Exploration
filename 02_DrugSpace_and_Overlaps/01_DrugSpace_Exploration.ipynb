```python
import pandas as pd
```

```python
ATC = pd.read_csv("../data/KeggDrugVirtualScreening/ATC_KEGG_processed.csv.gz", compression="gzip")
ATC
```

```python
JUMPCP_compounds = pd.read_csv("../data/processed/JUMPCP_compounds/JUMPCP_compounds_processed.csv.gz", compression="gzip")
JUMPCP_compounds = JUMPCP_compounds.drop_duplicates(subset="Standardized_SMILES", keep="first")
JUMPCP_compounds = JUMPCP_compounds.drop_duplicates(subset="Standardized_InChI", keep="first").reset_index(drop=True)
JUMPCP_compounds
```

```python
df = pd.merge(ATC, JUMPCP_compounds, on="Standardized_SMILES")
df = df.drop_duplicates(subset=["Standardized_SMILES", "BigGroup_ATC_class", "ATC_label_class"], keep="first").reset_index(drop=True)
df
```

```python
print(len(df.Standardized_SMILES.unique()))
```

```python
BigGroup_ATC_class_counts = pd.DataFrame(df["BigGroup_ATC_class"].value_counts())
BigGroup_ATC_class_counts
```

Check Distribution of BigFroup ATC


```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6), dpi=200) 

BigGroup_ATC_class_counts['BigGroup_ATC_class'].plot(kind='barh', color='green', edgecolor='black')
plt.xlabel('Count')
plt.ylabel('BigGroup ATC class')
plt.title('Distribution of BigGroup ATC class')
plt.tight_layout()
plt.show()
```

Check Distribution of ATC Class Labels of unique drugs in the dataset:

```python
ATC_label_counts = df[["SMILES", "ATC_label_class"]].drop_duplicates(keep="first")
ATC_label_counts = pd.DataFrame(ATC_label_counts["ATC_label_class"].value_counts())
ATC_label_counts = ATC_label_counts[ATC_label_counts["ATC_label_class"]>10]

# Set threshold for 'Others'
threshold = 15

# Create 'Others' category
others = ATC_label_counts[ATC_label_counts['ATC_label_class'] < threshold].sum()
ATC_label_counts = ATC_label_counts[ATC_label_counts['ATC_label_class'] >= threshold]
ATC_label_counts.loc['Others'] = others
```

```python
import seaborn as sns

# Using seaborn's style
sns.set_theme(style="whitegrid", palette="pastel")

# Increase the figure size
fig, ax = plt.subplots(figsize=(22, 22), dpi=200)

# Function to place the autopct values within the donut
def func(pct, allvalues): 
    absolute = int(pct / 100.*np.sum(allvalues)) 
    return "{:.1f}%".format(pct, absolute)

# Plot pie chart with legend and percentages inside the donut
wedges, texts, autotexts = ax.pie(ATC_label_counts['ATC_label_class'], 
                                  autopct=lambda pct: func(pct, ATC_label_counts['ATC_label_class']), 
                                  startangle=30, 
                                  wedgeprops=dict(width=0.3))

# Display a legend on the side
ax.legend(wedges, ATC_label_counts.index, title="ATC Label Class", loc="center left", 
          bbox_to_anchor=(1, 0, 0.5, 1), prop={'size': 20}, title_fontsize='20')

ax.set_title('Distribution of ATC Label Class', fontsize=20)

# Adjust the properties of the text for better visibility
plt.setp(autotexts, size=20, color="black")

plt.tight_layout()
plt.show()
```
