```python
import pandas as pd
```

```python
df = pd.read_csv("compounds.csv.gz", compression= "gzip")
df
```

```python
from pandarallel import pandarallel
pandarallel.initialize(progress_bar=True)
from rdkit import Chem

def inchi_to_smiles(value):

    try: return Chem.MolToSmiles(Chem.MolFromInchi(value))
    except: return "Cannot_do"
    
df["smiles"] = df.Metadata_InChI.parallel_apply(inchi_to_smiles)
df
```

```python
df = df[["Metadata_JCP2022", "smiles"]]
df.to_csv("JUMPCP_compounds.csv.gz", compression='gzip', index=False)
```
