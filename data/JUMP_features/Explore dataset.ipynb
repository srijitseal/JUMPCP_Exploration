```python
import pandas as pd
```

```python
df = pd.read_csv("JUMP_features.csv.gz", compression="gzip")
```

```python
df = df[["Standardized_SMILES"]]
df
```

```python
import numpy as np

# Generate 2000 random columns with values between 1 and 1000
num_columns_to_insert = 2000
random_columns = np.random.randint(1, 1001, size=(len(df), num_columns_to_insert))

# Create column names for the random columns
random_column_names = [f'Random_{i}' for i in range(1, num_columns_to_insert + 1)]

# Create a DataFrame with random columns
random_data = pd.DataFrame(random_columns, columns=random_column_names)

# Concatenate the random columns with the median_df DataFrame
median_df = pd.concat([df, random_data], axis=1)

median_df
```

```python
df = median_df.to_csv("JUMP_features.csv.gz", compression="gzip")
```
