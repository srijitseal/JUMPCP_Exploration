```python
import pandas as pd
import numpy as np
```

```python
drugbank= pd.read_csv("structure_links5_1_9.csv")
drugbank
```

```python
ATC_KEGG_DRUG_DATABASE= pd.read_csv("ATC_KEGG_DRUG_DATABASE.csv")
ATC_KEGG_DRUG_DATABASE
```

```python
Kegg_ATC_database_info= pd.read_csv("Kegg_ATC_database_info.csv")
Kegg_ATC_database_info
```

```python
ATC= pd.merge(ATC_KEGG_DRUG_DATABASE, Kegg_ATC_database_info, left_on="KEGG_code", right_on="KEGG_code")
ATC = ATC[['SMILES', 'KEGG_code', 'CompoundName', 'BigGroup_ATC_class', 'ATC_label_class']]
ATC = ATC.drop_duplicates().reset_index(drop=True)
ATC
```

```python
len(ATC.SMILES.unique())
```

```python
import sys
sys.path.append('../../')
from scripts.standardise_smiles import standardize_jumpcp, inchi_from_standardised_smile, check_carbon
```

```python
# Apply the standardize_oasis function
ATC['Standardized_SMILES'] = ATC["SMILES"].parallel_apply(standardize_jumpcp)
# Convert standardized SMILES to InChI
ATC['Standardized_InChI'] = ATC['Standardized_SMILES'].parallel_apply(inchi_from_standardised_smile)
#Flag ompounds without carbon as cannot_do
ATC['Standardized_SMILES'] = ATC['Standardized_SMILES'].parallel_apply(check_carbon)
# Filter out SMILES strings that couldn't be standardized
ATC = ATC[ATC['Standardized_SMILES'] != "Cannot_do"]
# Filter out InChI strings that couldn't be standardized
ATC = ATC[ATC['Standardized_InChI'] != "Cannot_do"].reset_index(drop=True)
ATC
```

```python
drugbank['Standardized_SMILES'] = drugbank["SMILES"].parallel_apply(standardize_jumpcp)
drugbank['Standardized_InChI'] = drugbank['Standardized_SMILES'].parallel_apply(inchi_from_standardised_smile)
drugbank['Standardized_SMILES'] = drugbank['Standardized_SMILES'].parallel_apply(check_carbon)
drugbank = drugbank[drugbank['Standardized_SMILES'] != "Cannot_do"]
drugbank = drugbank[drugbank['Standardized_InChI'] != "Cannot_do"].reset_index(drop=True)
drugbank
```
